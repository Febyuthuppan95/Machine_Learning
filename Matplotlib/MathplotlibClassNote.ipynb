```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Day=[1,2,3,4,5,6,7]
Temp=[50,51,52,48,47,49,46]
```

```python
plt.plot(Day,Temp)
plt.show()
```

```python
plt.plot(Day,Temp,color='green',linestyle='dashed',marker='o',linewidth=2)
plt.show()
```

```python
plt.plot(Day,Temp,color='green',linestyle='dashdot',marker='o',linewidth=2)
plt.show()
```

```python
plt.title('Weather Info',fontsize=20)
plt.xlabel('Day',fontsize=15)
plt.ylabel('Temperature',fontsize=15)
plt.plot(Day,Temp,color='green',linestyle='dashed',marker='o',linewidth=2)
plt.show()
```

```python
# alpha is used for decreasing colour
plt.title('Weather Info',fontsize=20)
plt.xlabel('Day',fontsize=15)
plt.ylabel('Temperature',fontsize=15)
plt.plot(Day,Temp,color='green',linestyle='dashed',marker='o',linewidth=2,alpha=0.2)
plt.show()
```

```python
# Legend grid
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.title('Weather Info',fontsize=20)
plt.xlabel('Day',fontsize=15)
plt.ylabel('Temperature',fontsize=15)
plt.plot(days,max_t,marker='o',label='Max')
plt.plot(days,min_t,marker='o',label='Min')
plt.plot(days,avg_t,marker='o',label='Avg')
plt.legend(loc='best',shadow=True,fontsize="small")
plt.grid()
plt.show()
```

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,2,34,12]
```

```python
plt.title("Revenue Models for Top Companies")
plt.xlabel("Major Company")
plt.ylabel("Profit , Revenue")
plt.bar(company,revenue,label="Revenue")
plt.bar(company,profit,label="Profit")
plt.legend()
plt.show()
```

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.xticks(xpos,company)
plt.title("Revenue Models for Top Companies")
plt.xlabel("Major Company")
plt.ylabel("Profit , Revenue")
plt.bar(xpos-0.2,revenue,width=0.4,label="Revenue")
plt.bar(xpos+0.2,profit,width=0.4,label="Profit")
plt.legend()
plt.show()
```

```python
plt.barh(xpos,company)
```

```python
# Box Plot
# -> random.normal is used for creating normal distribution
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
np.random.seed(10)
data = np.random.normal(100,20,200)
```

```python
data
```

```python
plt.hist(data,rwidth=0.8)
```

```python
plt.figure(figsize=(10,5))
plt.boxplot(data)
plt.show()
```

```python
np.random.seed(10)

data_1 = np.random.normal(100, 10, 200)
data_2 = np.random.normal(90, 20, 200)
data_3 = np.random.normal(80, 30, 200)
data_4 = np.random.normal(70, 40, 200)

data = [data_1, data_2, data_3, data_4]
plt.figure(figsize=(10,7))
plt.boxplot(data)
plt.show()
```

```python
# VERTICAL 
np.random.seed(10)

data_1 = np.random.normal(100, 10, 200)
data_2 = np.random.normal(90, 20, 200)
data_3 = np.random.normal(80, 30, 200)
data_4 = np.random.normal(70, 40, 200)

data = [data_1, data_2, data_3, data_4]
plt.figure(figsize=(10,7))
plt.boxplot(data,vert=False,patch_artist=True)
plt.show()
```
