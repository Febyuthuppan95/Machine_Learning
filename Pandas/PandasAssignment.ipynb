```python
import os 
import pandas as pd
path = "E:\AI\LibrariesForAI\Pandas\data"
inputPathFile = "supermarket_sales.csv"
# data = pd.read_csv("data/abcd.csv")
def summaryOfData(path,inputPathFile):
    pathandFileName = path+"\\"+inputPathFile
    extension = os.path.basename(pathandFileName).split(".")[1]  
    if extension=="csv":
        data = pd.read_csv(pathandFileName)
    elif extension == "xlsx":
        data = pd.read_excel(pathandFileName)
    elif extension == "xls":
        data = pd.read_excel(pathandFileName)
    elif extension == "json":
        data = pd.read_json(pathandFileName)
    else:
        print("Extension not found, Please recheck")
    print("The shape of the input data is : \n\n ",data.shape)
    print("*"*100+"\n")
    print("Data type of the input data is :\n\n ",data.dtypes)
    print("*"*100+"\n")
    print("Head of input data is : \n\n",data.head(2))
    print("*"*100+"\n")
    print("Tail of input data is : \n\n",data.tail(2))
    print("*"*100+"\n")  
    print("The basic Information about input data : \n\n")
    print(data.info())
    print("*"*100+"\n")
    print("Describe the input data : \n\n",data.describe())
    print("*"*100+"\n") 
    print("Describe about object in the input data : \n\n",data.describe(include='O'))
    print("*"*100+"\n")
    print("Null values in the input object : \n\n",data.isna().sum())
    print("*"*100+"\n")
summaryOfData(path,inputPathFile)
    
```
