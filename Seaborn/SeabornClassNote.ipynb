```python
# relationship we can perform different analysis
# -> 1 D Plot -> one feature -> univariate analysis
# 2 D Plot -> f1,f2 -> two feature -> bivariate analysis
# for multiple features -> multivariate analysis

# Displot  -> univariate analysis
# joinplot -> bivariate analysis
# Pairplot -> for more than 2 plot
```

```python
sns.jointplot(x='tip',y='total_bill',data=df)
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```
